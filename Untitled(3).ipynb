```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
```

```python
data = pd.read_csv('teamsGeolocation.csv')
df = pd.DataFrame(data)
print(data)
```

```python
df.plot(kind='scatter',
        x='Longitude',
        y='Latitude',)
plt.show()
```

```python
model = smf.quantreg('Latitude ~ Longitude', df).fit(q=0.5)
print(model.summary())
```

```python
northConf = []
southConf = []
for index, row in df.iterrows():
    new_row = {'City' : row['City'], 'Latitude' : row['Latitude'], 'Longitude' : row['Longitude']}
    if row['Latitude']>47.3757+0.0980*row['Longitude']:
        northConf.append(new_row)
    else :
        southConf.append(new_row)

dfNC = pd.DataFrame(northConf)
print("North Conference")
print("________________")
print(dfNC)
dfSC = pd.DataFrame(southConf)
print("South Conference")
print("________________")
print(dfSC)

       
        
```

```python
modelNorthConf = smf.quantreg('Latitude ~ Longitude', dfNC).fit(q=0.5)
print(modelNorthConf.summary())
```

```python
northNorthConf = []
northSouthConf = []
for index, row in dfNC.iterrows():
    new_row = {'City' : row['City'], 'Latitude' : row['Latitude'], 'Longitude' : row['Longitude']}
    if row['Latitude']>48.6792+0.0924*row['Longitude']:
        northNorthConf.append(new_row)
    else :
        northSouthConf.append(new_row)

dfNNC = pd.DataFrame(northNorthConf)
print("North North Conference")
print("________________")
print(dfNNC)
dfNSC = pd.DataFrame(northSouthConf)
print("North South Conference")
print("________________")
print(dfNSC)
```
