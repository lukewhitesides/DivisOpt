```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
```

```python
data = pd.read_csv('teamsGeolocation.csv')
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
model = smf.quantreg('Latitude ~ Longitude', df).fit(q=0.5)
print(model.summary())
```

```python
latMed=df['Latitude'].median()
print(latMed) 
longMed=df['Longitude'].median()
print(longMed)
df.loc[31]=["Combined Median",latMed,longMed]
print("_______")
print(df)
```

```python
df.plot(kind='scatter',
        x='Longitude',
        y='Latitude',)
plt.show()
```
